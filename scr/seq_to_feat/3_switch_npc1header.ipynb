```python
import pandas as pd
import os
import re
```

```python
# user options
switch_mode = 2 # either 1, 2
cutoff = 15
# switch_mode 1: convert specie name to Bx
# swith_mode 2: convert specie name to Bx|specie name
rootdir = '/Users/gorkalasso/Documents/GitHub/filo_GP-bat_NPC1/scr/seq_to_feat/'
infile = rootdir + 'output/2_interface_2_fasta/' +  str(cutoff) + '/interface_host_' + str(cutoff) + '.txt'
# infile = rootdir + str(cutoff) + '/interface_host_' + str(cutoff) + '_nr100'
# infile = rootdir + str(cutoff) + '/interface_host_' + str(cutoff) + '_nr100.clstr'
# infile = '/Users/gorkalasso/Dropbox (EinsteinMed)/ebola/NPC1_GP1_binding_determinants/1_NPC1_seqs/4_ali_on_5f1bC_1.fasta'
```

```python
# setting output file
filename, ext = os.path.splitext(infile)
# print(filename, ext)
tag = ''
if switch_mode == 1:
    tag = 'Bx'
elif switch_mode == 2:
    tag = 'Bx_sp'
outfile = filename + '_' + tag + ext
print(outfile)
```

```python
# load NPC1 Mapping table
mapfile = rootdir + 'input/0_npc1_list.xlsx'
sheetnm = 'in_project'
df = pd.read_excel(mapfile, sheet_name = sheetnm)
# display(df)
```

```python
infl = open(infile, 'r')
outfl = open(outfile,'w')
```

```python
# get format of input
fmt = 'fasta'
for ln in infl:
    ln = ln.rstrip()
    if len(ln) > 8 and ln[0:8] == '>Cluster':
        fmt = 'cluster'
print('Input format is', fmt)
```

```python
infl.seek(0)
if fmt == 'cluster':
    for ln in infl:
        ln = ln.rstrip()
        print(ln)
        if '>Cluster' in ln:
            print(ln, file=outfl)
        else:
            t1, t2 = ln.split('>')
            sp,t3 = t2.split('...')
            sp = re.sub('\s','',sp)
            sp = re.sub('\_',' ',sp)
            sp = re.sub('sp$','sp.',sp)
            bx = ''
            print('#' + sp + '#')
            if df['Species'].str.contains(sp).any():
#                 print('\t', 'found!!')
                idx = df.index[df['Species'] == sp].tolist()[0]
                bx = df.at[idx,'Label']
            elif sp == 'NPC1 human':
                bx = 'DC52'
            else:
                print('Fatal error, couldnt find BX for', sp)
                quit()
            if switch_mode == 1:
                newnm = bx
            elif switch_mode == 2:
                newnm = bx + '|' + sp
            print(t1 + '>' + newnm + '...' + t3, file=outfl)
elif fmt == 'fasta':
    for ln in infl:
        ln = ln.rstrip()
        print(ln)
        if ln[0:1] != '>':
            print(ln, file=outfl)
        else:
            sp = ln[1:]
            sp = re.sub('\s','',sp)
            sp = re.sub('\_',' ',sp)
            sp = re.sub('sp$','sp.',sp)
            bx = ''
            print('#' + sp + '#')
            if df['Species'].str.contains(sp).any():
#                 print('\t', 'found!!')
                idx = df.index[df['Species'] == sp].tolist()[0]
                bx = df.at[idx,'Label']
            elif sp == 'NPC1 human':
                bx = 'DC52'
            else:
                print('Fatal error, couldnt find BX for', sp)
                quit()
            if switch_mode == 1:
                newnm = bx
            elif switch_mode == 2:
                newnm = bx + '|' + sp
            print('>' + newnm, file=outfl)
```

```python
outfl.close()
infl.close()
```
